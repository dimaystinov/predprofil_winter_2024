```python
import cv2

img = cv2.imread('1.jpg') # считать изображение

# cv2.imshow('img', img) # показать изображение

shape = img.shape # высота, ширина, цветность
print(type(img))

a = round(min(shape[0],shape[1]) * 0.52)
height, width = shape[0],shape[1]
#print((height - a // 2), (height + a // 2))
img[(height // 2 - a // 2):(height // 2 + a // 2), (width // 2 - a // 2):(width // 2 + a // 2)] = [170,0,255] # строки, столбцы
cv2.imshow('img', img)
print(shape)
# print(img)
cv2.waitKey(0) # ожидать нажатия клавиши 0 мс
```

```python
cv2.destroyAllWindows() # закрыть все окна
```

```python
import cv2

capture = cv2.VideoCapture(0) # захватить видео с нулевой (встроенной камеры)

while True:
    ret, frame = capture.read() # флаг успешности захвата, кадр
    print(frame.shape)
    h, w, _ = frame.shape
    
    cv2.imshow('img',frame) 
    
    for i_str in range(h):
        frame[i_str] = frame[i_str][::-1]
        
    cv2.imshow('img_mirror',frame)
    key = cv2.waitKey(1) # Ждать 1 мс нажатия клавиши
    
    if key == ord(' '): #Клавиша пользователя - пробел
        break
        
capture.release() # Освободить камеру
cv2.destroyAllWindows() # Уничтожить все окна cv2
    
    
```
